```
"""
날짜 : 2020/09/08
이름 : 유효진
내용 : 한글 Word2vec 실습하기
"""
import codecs
from konlpy.tag import Okt
from gensim.models import word2vec
```

```
# 데이터셋 로드
f = codecs.open('/content/drive/My Drive/Tensorflow_works/data/ratings_train.txt', mode='r', encoding='utf-8')
lines = f.read().splitlines()
f.close()
```

```
# 데이터 확인 (tsv : tab으로 구분)
lines
```

```
# 데이터 전처리
ratings_train = []

for line in lines:
  data = line.split('\t')

  ratings_train.append(data)

# header 정보 행 제거
ratings_train = ratings_train[1:]

docs = []
for line in ratings_train:
  docs.append(line[1])

print(docs[0])
print(docs[1])
print(docs[99])
```

```
# 데이터 형태소 분석
okt = Okt()

results = []

for doc in docs:
  okt_pos = okt.pos(doc, norm=True, stem=True)
  #print(okt_pos)
  row = []

  for word in okt_pos:
    row.append(word[0])

  results.append(row)

print(results)
```

```
# word2vec 모델 생성
model = word2vec.Word2Vec(results)
model
```

```
# 단어 연관성 확인
rs1 = model.wv.most_similar(positive=['남자', '여배우'],
                             negative=['배우'])

rs2 = model.wv.most_similar(positive=['정우성', '조인성'])

print('rs1 :', rs1)
print('rs2 :', rs2)
```
